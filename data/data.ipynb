```python
# Import Section
from altair import Chart
from pandas import read_csv
from seaborn import pairplot
import matplotlib.pyplot as plt
```

# Loading Data

```python
# Loading the data
df = read_csv("rent.csv")
df.head()
```

# Data Visualization 

```python
Chart(df).mark_point().encode(x='Rent', y='Size',shape='Borough')
```

```python
pairplot(df, hue='Borough', height=5)
plt.show()
```
